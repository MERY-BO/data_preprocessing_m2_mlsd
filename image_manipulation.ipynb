```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import random
from PIL import Image
import numpy as np
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("msarmi9/food101tiny")

print("Path to dataset files:", path)
```

```python
# Path to the class folder (e.g., cannoli)
data_path = os.path.join(path, 'data','food-101-tiny','train','cannoli')
image_files = os.listdir(data_path)
```

```python
# 3) Visualize first 5 images ---
plt.figure(figsize=(12, 6))
for i, fname in enumerate(image_files[:5]):
    img_path = os.path.join(data_path, fname)
    img = mpimg.imread(img_path)
    plt.subplot(1, 5, i+1)
    plt.imshow(img)
    plt.axis("off")
    plt.title(f"img {i+1}")
plt.show()
```

```python
# 4) Normalize pixels to [0,1] -
# Take the second image
img2_path = os.path.join(data_path, image_files[1])
img2 = mpimg.imread(img2_path).astype(np.float32)

# If values are in [0,255], normalize
if img2.max() > 1.0:
    img2 = img2 / 255.0
# 2) shape of the image
print("Shape of 2nd image:", img2.shape)
print("Pixel range:", img2.min(), "to", img2.max())

# 5) Calculate mean & std of pixels ---
mean_val = img2.mean()
std_val = img2.std()

print(f"Mean pixel value: {mean_val:.4f}")
print(f"Std pixel value: {std_val:.4f}")
```
